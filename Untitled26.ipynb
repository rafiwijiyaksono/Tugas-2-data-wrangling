```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/GeostatsGuy/GeoDataSets/master/1D_Porosity.csv")

df.head()
```

```python
X = df[['Depth']]
y = df['Nporosity']
```

```python
model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)
```

```python
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE :", rmse)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred)

plt.xlabel("Depth")
plt.ylabel("Porosity")
plt.title("Regresi Linear Depth vs Porosity")

plt.show()
```

```python
for i in range(2,6):

    poly = PolynomialFeatures(degree=i)
    X_poly = poly.fit_transform(X)

    model = LinearRegression()
    model.fit(X_poly, y)

    y_pred = model.predict(X_poly)

    plt.scatter(X, y)
    plt.plot(X, y_pred)

    plt.title("Polynomial Regression Orde " + str(i))
    plt.xlabel("Depth")
    plt.ylabel("Porosity")

    plt.show()
```

Pada praktikum ini dilakukan analisis hubungan antara variabel Depth dan Porosity menggunakan metode regresi linear dan regresi polynomial. Regresi linear digunakan untuk memodelkan hubungan sederhana antara dua variabel dengan menggunakan garis lurus sebagai pendekatan terbaik terhadap data.

Berdasarkan model yang dibuat, dilakukan evaluasi menggunakan beberapa metrik error yaitu MAE (Mean Absolute Error), MSE (Mean Squared Error), dan RMSE (Root Mean Squared Error). Ketiga metrik ini digunakan untuk mengukur seberapa baik model dalam melakukan prediksi terhadap data yang ada. Nilai error yang lebih kecil menunjukkan bahwa model memiliki tingkat kesalahan yang lebih rendah dan mampu memprediksi data dengan lebih baik.

Selain regresi linear, dilakukan juga analisis menggunakan polynomial regression dengan beberapa tingkat orde yaitu orde 2, 3, 4, dan 5. Tujuan dari penggunaan polynomial regression adalah untuk melihat apakah hubungan antara variabel Depth dan Porosity memiliki pola yang lebih kompleks yang tidak dapat sepenuhnya dijelaskan oleh model linear.

Dari grafik yang dihasilkan dapat dilihat bahwa semakin tinggi orde polynomial yang digunakan, maka kurva yang dihasilkan menjadi semakin fleksibel dan mampu mengikuti pola data dengan lebih detail. Namun penggunaan orde yang terlalu tinggi juga dapat menyebabkan overfitting, yaitu kondisi di mana model terlalu menyesuaikan data yang ada sehingga kurang baik dalam melakukan generalisasi terhadap data baru.

Secara keseluruhan, praktikum ini menunjukkan bahwa pemilihan model regresi yang tepat sangat penting dalam analisis data. Dengan menggunakan regresi linear dan polynomial regression, kita dapat memahami pola hubungan antara variabel serta mengevaluasi performa model menggunakan metrik error yang sesuai.
